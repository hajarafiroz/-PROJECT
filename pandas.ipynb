```python
import pandas as pd
# From dictionary
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)
print(df)
```

```python
# From list of lists
lst = [['Alice', 25], ['Bob', 30]]
df = pd.DataFrame(lst, columns=['Name', 'Age'])
print(df)
```

```python
df1 = pd.DataFrame({
    'key': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'John']
})
```

```python
df2 = pd.DataFrame({
    'key': [1, 2, 3],
    'Age': [20, 21, 22]
})
```

```python
# Merge two DataFrames
pd.merge(df1, df2, on='key')
```

```python
# Join by index
df1.set_index('key').join(
    df2.set_index('key')
)
```
